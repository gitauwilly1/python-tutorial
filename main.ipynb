```python
student = {
    'name': 'Alice',
    'age': 20,
    'course': 'Python Programming'
}

print(student)
print(student['name']) # Accessing the value associated with the key 'name'
print(student.get('city')) # Accessing the value associated with the key 'city' using the get method
```

```python
for key in student:
    print(key, student[key]) # Printing each key and its corresponding value in the dictionary
```

```python
for key, value in student.items():
    print(key, value) # Printing each key and its corresponding value using the items() method
```

```python
import requests

url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url) 

if response.ok: # Checking if the request was successful using the 'ok' property of the response object
    print('Request was successful!')
    print(response.json())

else:
    print("Request failed with status code:", response.status_code)
```

```python
import requests

url = "https://jsonplaceholder.typicode.com/posts"

try:
    response = requests.get(url) 

    if response.ok: 
        print(response.json())

    else:
        print("Error", response.status_code)

except requests.exceptions.RequestException as e:
    print("An error occurred while making the request:", e)
```

```python
import requests

url = "https://jsonplaceholder.typicode.com/posts"

try:
    response = requests.get(url) 

    if response.ok: 
        posts = response.json() # Storing the JSON data from the response in a variable named 'posts'

        for post in posts:
            print(f"Title: {post['title']}") # Printing the title of each post from the JSON data
            print(f"Body: {post['body']}") # Printing the body of each post from the JSON data
    else:
        print("Error", response.status_code)

except requests.exceptions.RequestException as e:
    print("An error occurred while making the request:", e)
```
