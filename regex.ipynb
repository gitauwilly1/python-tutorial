```python
import re
txt = "The rain in Spain stays mainly in the plain."

match = re.search(r"ain", txt) # Search for the first occurrence of "ain" in the text

if match:
    print("Match found:", match.group())
else:
    print("No match found.")
```

```python
text = "Hello world!"
result = re.match(r"Hello", text) # Check if the text starts with "Hello"
print(result)  
```

```python
txt = "The rain in Spain stays mainly in the plain."
words = re.findall(r"ain", txt) # Find all words that end with "ain"
print(words)
```

```python
# . matches any character except a newline

text = 'cat cot cut cit'
result = re.findall(r'c.t', text)
print(result)
```

```python
# \d matches any digit

text = 'My phone number is 123-456-7890.' 
result = re.findall(r'\d', text)
print(result)
```

```python
# \w matches any alphanumeric character (letters, digits, and underscores)

text = 'The price is $19.99.'
result = re.findall(r'\w', text)  
print(result)
```

```python
# \s matches any whitespace character (spaces, tabs, newlines)

text = 'Hello world! This is a test.'
result = re.findall(r'\s', text)
print(result)
```

```python
# ^ matches the start of a string

text = "Hello world!"
result = re.findall(r'^Hello', text) # Check if the text starts with "Hello"
print(result)
```

```python
# $ matches the end of a string

text = "Hello world!"
result = re.findall(r'world!$', text) # Check if the text ends with "world!"
print(result)
```

```python
# {n} matches exactly n occurrences of the preceding character or group

text = '123, 1234, 12345, 123456'
result = re.findall(r'\d{4}', text) # Find all sequences of exactly 4 digits
print(result)
```

```python
# [] matches any single character within the brackets

text = 'cat bat cot cut cit pot'
result = re.findall(r'[cb]', text) # Find all occurrences of 'c' or 'b'
print(result)
```

```python
# | matches either the pattern before or the pattern after the | symbol

text = 'I like cats and dogs.'
result = re.findall(r'cats|dogs', text) # Find all occurrences of "cats" or "dogs"
print(result)
```

```python
# \D matches any non-digit character

text = 'My phone number is 123-456-7890.'
result = re.findall(r'\D', text) # Find all non-digit characters
print(result)
```

```python
# \W matches any non-alphanumeric character (anything that is not a letter, digit, or underscore)

text = 'The price is $19.99.'
result = re.findall(r'\W', text) # Find all non-alphanumeric characters
print(result)
```

```python
# form validation

email = "user@example.com"
email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$' # Simple email pattern
if re.match(email_pattern, email):
    print("Valid email address.")
else:
    print("Invalid email address.")
```

```python
# password validation

password = "Password123!"
password_pattern = r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*?&])[A-Za-z\d@$!%*?&]{8,}$' # Complex password pattern
if re.match(password_pattern, password):
    print("Strong password.")
else:
    print("Weak password.")
```

```python
import re

def is_alphanumeric(text):
   
    pattern = re.compile(r'^[a-zA-Z0-9]+$')
    
    if pattern.match(text):
        print(f"'{text}' contains only allowed characters.")
    else:
        print(f"'{text}' contains invalid characters.")


is_alphanumeric("Python3")     
is_alphanumeric("Python 3!")   
```

```python
import re

def match_a_to_b(text):
  
    pattern = re.compile(r'^a.*b$')
    
    if pattern.match(text):
        print(f"'{text}' is a match!")
    else:
        print(f"'{text}' is NOT a match.")


match_a_to_b("ab")             # Match (zero characters in between)
match_a_to_b("apple_grab")     # Match
match_a_to_b("a12345b")        # Match
match_a_to_b("banana_grab")    # No match (doesn't start with 'a')
match_a_to_b("apple_grabbed")  # No match (doesn't end with 'b')
```
