```python
import requests
from bs4 import BeautifulSoup 

page = 1
while True:

    response = requests.get(f'https://books.toscrape.com/catalogue/page-{page}.html')
    soup = BeautifulSoup(response.text, 'html.parser')
    # print(soup)
    books = soup.select('.product_pod')
    # print(books)

    if not books:
        break

    for book in books:
        title = book.h3.a['title']
        rating_map = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
        rating = rating_map.get(book.p['class'][1], 'No rating')
    price = book.find('p', class_='price_color').text
    print(f'Title: {title}\nRating: {rating}\nPrice: {price}\n')

    page += 1
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup 

page = 1
while True:

    response = requests.get(f'https://books.toscrape.com/catalogue/page-{page}.html')
    soup = BeautifulSoup(response.text, 'html.parser')
    # print(soup)
    books = soup.select('.product_pod')
    # print(books)

    if not books:
        break

    for book in books:
        title = book.select_one('h3 a')['title']
        rating_map = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
        rating = rating_map.get(book.p['class'][1], 'No rating')
        price = book.select_one('.price_color')['text']
        stock = book.select_one('.availability').text.strip()
        print(f'Title: {title}\nRating: {rating}\nPrice: {price}\nStock: {stock}\n')

        book_item = {
            'title': title,
            'rating': rating,
            'price': price,
            'stock': stock
        }

        book

    df = pd.DataFrame([book_item])
    print(df)

    page += 1
```
